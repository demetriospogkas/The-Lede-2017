## Request the page with the correct data parameters AND the correct MINIMUM headers

This time it should work.

```python
url_barbers = "https://www.tdlr.texas.gov/cimsfo/fosearch_results.asp"

data = {
"pht_status":"BAR",
"pht_lic":"",
"pht_lnm":"",
"pht_fnm":"",
"pht_oth_name":"",
"phy_city":"HOUSTON",
"phy_cnty":"-1",
"phy_zip":"",
"B1":"Search"
}

headers = {
    'Referer' : 'https://www.tdlr.texas.gov/cimsfo/fosearch.asp'
}
response_barbers = requests.post(url_barbers, data = data, headers = headers)
doc_barbers = BeautifulSoup(response_barbers.text, "html.parser")
doc_barbers
```

## Scraping

### Loop through each `tr` and print each person's name

You'll get an error because the first one doesn't have a name. How do you make that not happen? I'm happy to help if you ask on the board.

```python
trs = doc_barbers.find_all('tr')
trs
```

```python
for tr in trs[1:]:
    name = tr.find_all('td')[0].find_all('span')[0]
    name = name.string.title()
    print(name)
```

## Loop through each `tr`, printing each violation description

- TIP: What is the container tag name for it?
- TIP: You'll get an error even if you're ALMOST right - which row is causing the problem?

```python
for tr in trs[1:]:
    violation = tr.find_all('td')[2]
    violation = violation.string
    print(violation)
```

## Loop through each `tr`, printing the complaint number

- TIP: It should be the last piece of the fist `td`

```python
for tr in trs[1:]:
    complaint_no = tr.find_all('td')[0].find_all('span')[-1]
    complaint_no = complaint_no.string.strip()
    print(complaint_no)
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Person's name
- Violation description
- Violation number

Create a new dictionary for each `tr` (except the header).

```python
license_violation_barbers = []

for tr in trs[1:]:
    barber_dict = {}
    barber_name = tr.find_all('td')[0].find_all('span')[0].string.title()
    barber_violation = violation = tr.find_all('td')[2].string
    barber_complaint_no = complaint_no = tr.find_all('td')[0].find_all('span')[-1].string.strip()
    barber_dict['Barber_name'] = barber_name
    barber_dict['Barber_violation'] = barber_violation
    barber_dict['Barber_complaint_no'] = barber_complaint_no
    license_violation_barbers.append(barber_dict)

print(len(license_violation_barbers))
print(license_violation_barbers)
```

### Save that to a CSV

```python
df_barbers = pd.DataFrame(license_violation_barbers)
df_barbers.shape
```
