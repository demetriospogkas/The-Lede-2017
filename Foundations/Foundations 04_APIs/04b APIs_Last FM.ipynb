## 12) Using the first ten mbids, print the artist's name and whether they're a rapper

Only print their name ONCE and only print whether they are hip hop or not ONCE.

- *Tip: Rap tags include hip hop, swag, crunk, rap, dirty south, and probably a bunch of other stuff! You can include as many categories as you'd like.*
- *Tip: Every time you look at a new artist, you can say they are NOT a rapper. And once you find out one of their tags is hip hop or rap, then you can note that they're a rapper. Then once you're done looking at their tags, then you can say HEY this is a rapper, or HEY this is not a rapper.*

````python
import re

print("Our 10 first artists, and their names:")

for ttr in artist_mbids[:10]:
    url_ten = "http://ws.audioscrobbler.com/2.0/?method=artist.getinfo&api_key={YOUR_API_KEY}&format=json&mbid=" + ttr
    response_ten = requests.get(url_ten)
    data_ten = response_ten.json()
    results_ten = data_ten['artist']
    name_ten = results_ten['name']
    print("ARTIST:", name_ten)
    tags_ten = results_ten['tags']['tag']
    tags_lists = [mko['name'] for mko in tags_ten]
    tags_rap = []
    for qwer in tags_lists:
        re.sub('[^a-z]', '', qwer.lower())
        match = re.search(r'(rap)|(Hip-Hop)', qwer)
        if match == None:
            tags_rap.append("no rap")
        else:
            tags_rap.append("rap")
    tags_rap = sorted(tags_rap)
    for erty in tags_rap:
        if erty == "rap":
            rap_verdict = "YES hip hop"
        else:
            rap_verdict = "NO hip hop"
    print(rap_verdict)
````

## 13) What percent of "lil" results are rappers?

````python
rappers = []
non_rappers = []

for ttr in artist_mbids[:50]:
    url_ten = "http://ws.audioscrobbler.com/2.0/?method=artist.getinfo&api_key={YOUR_API_KEY}&format=json&mbid=" + ttr
    response_ten = requests.get(url_ten)
    data_ten = response_ten.json()
    results_ten = data_ten['artist']
    tags_ten = results_ten['tags']['tag']
    tags_lists = [mko['name'] for mko in tags_ten]
    tags_rap = []
    for qwer in tags_lists:
        re.sub('[^a-z]', '', qwer.lower())
        match = re.search(r'(rap)|(Hip-Hop)', qwer)
        if match == None:
            tags_rap.append("no rap")
        else:
            tags_rap.append("rap")
    tags_rap = sorted(tags_rap)
    for erty in tags_rap:
        if erty == "rap":
            rap_verdict = "YES hip hop"
        else:
            rap_verdict = "NO hip hop"
    if rap_verdict == "YES hip hop":
        rappers.append(1)
    else:
        non_rappers.append(1)

sum_rappers = sum(rappers)
sum_non_rappers = sum(non_rappers)

pc_rappers = sum_rappers / (sum_rappers + sum_non_rappers)
pc_rappers = round(pc_rappers*100, 2)

pc_non_rappers = sum_non_rappers / (sum_rappers + sum_non_rappers)
pc_non_rappers = round(pc_non_rappers*100, 2)

print("In our results we have", pc_rappers, "% rappers, and", pc_non_rappers, "% non rappers.")
````
