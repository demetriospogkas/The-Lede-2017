```python
import pandas as pd
```

```python
# Read our CSV
work_fatalities_2015 = pd.read_csv('fy15_federal-state_summaries.csv', error_bad_lines=True)
work_fatalities_2015
```

```python
# Let's re-open our CSV, delete unnecessary rows and columns and read it again
work_fatalities_2015 = pd.read_csv('fy15_federal-state_summaries_ending-lines.csv', encoding = "ISO-8859-1")
work_fatalities_2015
```

```python
# Better now
work_fatalities_2015.tail()
```

```python
# Let's separate year and months
work_fatalities_2015['year'] = work_fatalities_2015['Date of Incident'].str.extract('[/](\d{4})')
work_fatalities_2015['month'] = work_fatalities_2015['Date of Incident'].str.extract('(.*)[/].*[/]')
work_fatalities_2015
```

```python
# Let's have the month in letters too
# Built a CSV with the months table in: https://en.wikipedia.org/wiki/Month
months = pd.read_csv('months_Wikipedia-csv.csv')
months
```

```python
work_fatalities_2015.merge(months, left_on='month', right_on='Chronology')
# Doesn't seem to work
```

```python
# Let's see if they're the same dtypes
months.dtypes
```

```python
work_fatalities_2015.dtypes
```

```python
# They aren't, so let's turn Chronology to string
months['Chronology'] = months['Chronology'].astype(str)
months.dtypes
```

```python
# Let's see if the merge works now
work_fatalities_2015 = work_fatalities_2015.merge(months, how='left', left_on='month', right_on='Chronology')
work_fatalities_2015
```

```python
# It worked! But we don't care for this Days column, let's drop it.
work_fatalities_2015.drop(['Days', 'Chronology'], axis=1, inplace=True)
work_fatalities_2015
```

```python
# Let's graph what year had the most fatalities
%matplotlib inline
work_fatalities_2015.year.value_counts().plot(kind='bar')
```

```python
# Let's graph which months had the most fatalities
work_fatalities_2015.Alphabetic.value_counts(ascending=True).plot(kind='barh')
```

```python
# Let's see if any company had multiple working fatalities
# We will keep in our extract the name of the city, in case there are companies with the same name in other cities
work_fatalities_2015['company'] = work_fatalities_2015['Company, City, State, ZIP'].str.extract('^(.*),')
work_fatalities_2015.company.value_counts().head(10)
```

```python
# Let's extract the state in which are based the companies that their workers had accidents
work_fatalities_2015['state'] = work_fatalities_2015['Company, City, State, ZIP'].str.extract('([A-Z]{2}) \d{5}')
work_fatalities_2015.head(5)
```

```python
# Let's add the names of each state next to their codes
# First let's import a file with their names
# Built a CSV from the table in: https://www.census.gov/geo/reference/ansi_statetables.html
state_names = pd.read_csv('US-state-codes.csv')
state_names.head(10)
```

```python
work_fatalities_2015 = work_fatalities_2015.merge(state_names, left_on='state', right_on=' Official USPS Code')
work_fatalities_2015
```

```python
work_fatalities_2015.drop(['FIPS State Numeric Code', ' Official USPS Code'], axis=1, inplace=True)
work_fatalities_2015
```

```python
# Let's graph which states had the most fatalities
work_fatalities_2015.Name.value_counts(ascending=True).plot(kind='barh', figsize=(10,10))
```

```python
# Let's see which states had the most incidents per 100,000 residents
# First let's generate a count of incidents per state:
count_per_state =  work_fatalities_2015.groupby('Name')['Date of Incident'].agg(['count'])
count_per_state
```

```python
# Then let's import the 2010 US census data
# Built a CSV from the table found on: https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population
us_census = pd.read_csv('US-census-2010.csv')
us_census
```

```python
count_per_state = count_per_state.merge(us_census, how='left', left_index=True, right_on='State or territory')
count_per_state
```

```python
count_per_state['per_capita'] = (count_per_state['count'] / count_per_state['Census population-April-1-2010']) * 100000
count_per_state
```

```python
count_per_state = count_per_state.round(decimals=2)
count_per_state.sort_values(by='per_capita', ascending=False).head(10)
```

```python
count_per_state.plot(kind='scatter', x='Census population-April-1-2010', y='per_capita')
```
