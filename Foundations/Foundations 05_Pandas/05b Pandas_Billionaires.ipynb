## 8) How much money do these billionaires have in total?

```python
total_money = df['networthusbillion'].sum()
print("They have", total_money/1000, "trillion dollars")
```

```python
companies = ['Hyatt', 'Oetker-Gruppe', 'S. C. Johnson & Son', 'Walmart', 'Alfa Group']

total_sum = 0
for company in companies:
    df_companies=df[df['company'] == company]
    total_companies = df_companies['networthusbillion'].sum()
    total_sum = total_sum + df_companies['networthusbillion'].sum()
print(total_sum)
```

## 9) What are the top 3 countries with the most money held by billionaires?
